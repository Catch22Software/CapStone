```python
df['Transaction_ID'] = 'ID: '+ df['Member_number'] + ' Date: '+ df['Date']
df.head()
```

```python
df.drop(columns=['Member_number', 'Date'], inplace=True)
df.head()
```

```python
df.head()
```

```python
print(len(pd.unique(df['itemDescription'])))
```

```python
df['Transaction_ID'].is_unique
```

```python
transactions = defaultdict(list)
for index, row in df.iterrows():
    transactions[row['Transaction_ID']].append(row['itemDescription'])
print(len(transactions))
```

```python
print(transactions)
```

```python
print(transactions.values())
```

```python
total = 0
items = list(transactions.copy().values())
for i in range(len(items)):
    total += len(items[i])
print(total)
```

```python
print(items)
```

```python
file = open('Data/filtered_transaction_data.csv', 'w', newline='')
with file:
    write = csv.writer(file)
    write.writerows(items)
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit(items).transform(items)  # take transaction data and transform into df
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
print(len(df))  # length of initial hot-encoded pandas dataframe
```

```python
frequent_1 = apriori(df, min_support=0.000668315, use_colnames=True)  # support for at least 500 of the 14963 transactions
print(len(frequent_1))  # length of initial rules with 167 total unique items
frequent_1.head()
```

```python
rules_1 = association_rules(frequent_1, metric="confidence", min_threshold=0.5)
# adds a cutoff of at least 50% confidence
print(len(rules_1))  # length of resulting ruleset
```

```python
rules_1.consequents.unique()
print(len(rules_1.consequents.unique()))  # amount of unique consequent recommendations
```

```python
rules_1.to_csv('Data/master_rules.csv')
```
